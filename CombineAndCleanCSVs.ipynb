# Concatonate CSV Files

```python
#Make a collection containing all of the filenames of the csvs that were saved earlier
all_filenames=[]
for year in range(1935,2021):
    all_filenames.append(f"team_offense{year}.csv")
#Concatonate the csvs together using list comprehension and filenames list
#Each file in the collection will be read into and stored in a dataframe
#Skiprows=2 removes the headers I don't care about before writing to csv
combined_csv = pd.concat([pd.read_csv(f,skiprows=2) for f in all_filenames])

#Write the combined dataframe to csv
combined_csv.to_csv("combined_offense.csv")
```

# Clean Data

```python
combined_csv=pd.read_csv("combined_offense.csv")
```

```python
#This slice of the dataframe has all the years
year_col=combined_csv.iloc[:,28:]
year_col2=year_col.stack().dropna(axis=0)
#year_col.dropna(axis=0)
#I need to stack columns to consolidate and drop na values
year_col2["year"]=pd.DataFrame(year_col.stack().dropna(axis=0))

```

```python
#store the values as an array
years=year_col2["year"].values
```

```python
#Add the array containing the year values into the dataframe
combined_csv["year"]=years
```

```python
#I need to drop the columns that I combined into the year array.
#29 is where it starts 114 is the last col to drop so stop slice at 115. Save df slice as variable and then drop.
col_to_drop=combined_csv.iloc[:,29:115]
test_1=combined_csv.drop(col_to_drop,axis=1)
offense_df=test_1.drop(test_1.iloc[:,0:3],axis=1)
```

# Rename Columns

```python
#I need to rename columns to avoid ambiguity. The context is lost from the website headers.
offense_df.columns = ['Tm', 'G', 'PF', 'Total_Yds', 'Ply', 'Y/P', 'TO', 'FL', 'Total_1stD', 'Cmp', 'Pass_Att',
       'Pass_yds', 'Pass_TD', 'Int', 'NY/A', 'Pass_1stD', 'Rush_Att', 'Rush_Yds', 'Rush_TD', 'Y/A',
       'Rush_1stD', 'Pen', 'Penalty_Yds', 'Penalty_1stD', 'Sc%', 'TO%', 'Year'] 
```
